# 具体例①

```python
import pandas as pd
from prefixspan import PrefixSpan

# サンプルデータフレームの作成
data = {
    'TransactionID': [
        1, 1, 1,
        2, 2,
        3, 3,
        4,
        5, 5,
        6, 6, 6,
        7, 7,
        8, 8,
        9, 9, 9,
        10, 10,
        11, 11,
        12, 12, 12,
        13, 13,
        14, 14,
        15, 15, 15,
        16, 16,
        17, 17,
        18, 18,
        19, 19, 19,
        20, 20,
        21, 21,
        22, 22, 22,
        23, 23,
        24, 24,
        25, 25,
        26, 26, 26
    ],
    'Item': [
        'milk', 'bread', 'butter',
        'bread', 'butter',
        'milk', 'bread',
        'milk',
        'bread', 'butter',
        'bread', 'butter', 'eggs',
        'milk', 'eggs',
        'bread', 'butter',
        'bread', 'milk', 'butter',
        'milk', 'bread',
        'butter', 'eggs',
        'milk', 'bread', 'butter',
        'bread', 'butter',
        'eggs', 'milk',
        'bread', 'eggs', 'milk',
        'butter', 'bread',
        'butter', 'milk',
        'bread', 'butter',
        'eggs', 'milk', 'bread',
        'milk', 'butter',
        'bread', 'butter',
        'eggs', 'milk', 'bread',
        'bread', 'eggs',
        'milk', 'butter',
        'bread', 'butter',
        'bread', 'milk', 'butter'
    ]
}

df = pd.DataFrame(data)
display(df)

# トランザクションごとにアイテムをグループ化
grouped = df.groupby('TransactionID')['Item'].apply(list).tolist()

# PrefixSpanの実行
ps = PrefixSpan(grouped)

# 最小サポートを指定して頻出パターンを抽出
min_support = 2  # ここではサポートが2以上のパターンを抽出
patterns = ps.frequent(min_support)

# 結果の表示
print("頻出パターン:")
for pattern in patterns:
    display(pattern)
```

# 具体例②

```python
# サンプルのデータフレームを作成
data = {
    'user_id': [1, 1, 1, 2, 2, 3, 3, 3, 4, 4],
    'event': ['a', 'b', 'c', 'a', 'c', 'b', 'c', 'a', 'a', 'b'],
    'time': [1, 2, 3, 1, 2, 1, 2, 3, 1, 2]
}

df = pd.DataFrame(data)
display(df)

# ユーザーごとにイベントのリストを作成
sequences = df.groupby('user_id')['event'].apply(list).tolist()

# PrefixSpanを用いて頻出パターンを抽出
ps = PrefixSpan(sequences)

# 最小サポート数を指定（ここでは2）
min_support = 2  # ここではサポートが2以上のパターンを抽出
patterns = ps.frequent(min_support)

# 結果を表示
for pattern in patterns:
    print(f'パターン: {pattern[1]}, 出現回数: {pattern[0]}')
```
